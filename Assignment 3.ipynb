```python
import matplotlib.pyplot as plt
# Network structure
layers = [6, 8, 4, 2]
layer_names = ["Input Layer", "Hidden Layer 1", "Hidden Layer 2", "Output Layer"]
input_features = [
   "Transaction Amount",
   "Transaction Time",
   "Merchant Category",
   "Customer Age",
   "Account Balance",
   "No. of Transactions Today"
]
output_labels = ["Not Fraud (0)", "Fraud (1)"]
# Figure size
plt.figure(figsize=(16, 8))
x_spacing = 3
y_spacing = 1
# Draw neurons and connections
for i, layer_size in enumerate(layers):
   x = i * x_spacing
   y_start = -(layer_size - 1) / 2
   for j in range(layer_size):
       y = y_start + j * y_spacing
       plt.scatter(x, y, s=700)
       # Connections
       if i > 0:
           prev_layer_size = layers[i - 1]
           prev_y_start = -(prev_layer_size - 1) / 2
           for k in range(prev_layer_size):
               prev_y = prev_y_start + k * y_spacing
               plt.plot([x - x_spacing, x], [prev_y, y], linewidth=0.6)
       # Input feature labels
       if i == 0:
           plt.text(x - 2.3, y, input_features[j],
                    fontsize=10, ha='right', va='center')
       # Output labels (Separate Fraud & Not Fraud)
       if i == 3:
           plt.text(x + 0.8, y, output_labels[j],
                    fontsize=11, ha='left', va='center')
# GRAPH TITLE (TOP)
plt.title("Neural Network Architecture for Credit Card Fraud Detection",
         fontsize=15, fontweight='bold', pad=25)

# LAYER NAMES (BOTTOM)
for i, name in enumerate(layer_names):
   plt.text(i * x_spacing, -6,
            name, ha='center', fontsize=12, fontweight='bold')
plt.axis('off')
plt.tight_layout()
plt.show()
```
