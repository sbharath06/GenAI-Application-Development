```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
# Sample data (Assumed)
area = np.array([800, 1000, 1200, 1500, 1800, 2000])
bedrooms = np.array([1, 2, 2, 3, 3, 4])
price = np.array([30, 40, 50, 65, 75, 90])  # in Lakhs (example)
# Create mesh grid for regression plane
area_grid, bedroom_grid = np.meshgrid(
   np.linspace(area.min(), area.max(), 10),
   np.linspace(bedrooms.min(), bedrooms.max(), 10)
)
# Simple regression equation (for visualization)
price_grid = 0.04 * area_grid + 5 * bedroom_grid
# Create 3D plot
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')
# Scatter actual data points
ax.scatter(area, bedrooms, price,
          label="Actual House Data",
          s=60)
# Regression surface
ax.plot_surface(area_grid, bedroom_grid, price_grid,
               alpha=0.6)
# Axis labels
ax.set_xlabel("Area (Square Feet)", fontsize=11)
ax.set_ylabel("Number of Bedrooms", fontsize=11)
ax.set_zlabel("House Price (Lakhs)", fontsize=11)
# Title
ax.set_title("3D Regression Model Visualization for House Price Prediction",
            fontsize=14, fontweight='bold')
# Legend
ax.legend()
plt.show()
```
